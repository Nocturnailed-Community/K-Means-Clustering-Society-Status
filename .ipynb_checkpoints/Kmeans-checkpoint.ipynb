```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import joblib
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
#Menyiapkan data dan memanggil dataset
dataset = pd.read_csv('Data.csv')
dataset.keys()
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
#konversi ke data array
X = np.asarray(dataset)
print(X)
```

```python
#Menampilkan data dalam bentuk scatter plot
plt.scatter(dataset['JumlahAnak'],dataset['Penghasilan'], label='True Position')
plt.xlabel('JumlahAnak')
plt.ylabel('Penghasilan')
plt.title('Klustering Status Masyarakat')
plt.show()
```

```python
#Data Training
x_train = dataset[['JumlahAnak','Penghasilan']].values
x_train.view()
```

```python
#Clustering
kmean = KMeans(n_clusters=2)
y_cluster=kmean.fit_predict(x_train)
y_cluster
```

```python
#Field baru
dataset['Cluster']=y_cluster
dataset
```

```python
#Centroid
kmean.cluster_centers_
```

```python
#Visualisasi data
plt.figure(figsize=(10,5))
plt.scatter(x_train[:,0],x_train[:,1],c=kmean.labels_)
plt.scatter(kmean.cluster_centers_[0,0],kmean.cluster_centers_[0,1],cmap='rainbow',marker='*',s=150,label='Centroid 0')
plt.scatter(kmean.cluster_centers_[1,0],kmean.cluster_centers_[1,1],cmap='rainbow',marker='*',s=150,label='Centroid 1')
plt.xlabel('Jumlah Anak')
plt.ylabel('Penghasilan')
plt.title('Klustering Status Masyarakat')
plt.legend()
```

```python
#Mencari Jumlah Kluster Terbaik
inertias = []
k_range = range(1,10)
for k in k_range:
  km = KMeans(n_clusters=k).fit(x_train)
  inertias.append(km.inertia_)
```

```python
  inertias
```

```python
plt.xlabel('k')
plt.ylabel('sum of error')
plt.plot(k_range,inertias)
plt.grid()
```

```python
#Membuat Dataset Baru Dengan Label Cluster
dataset['Cluster'] = dataset['Cluster'].replace([0,1],['Miskin','Kaya'])
dataset
```

```python
#call model regression
model = kmean
#save model
filename = 'modelkmeans.sav'
joblib.dump(model, filename)
#load model
loaded_model = joblib.load(filename)
#prediction model
loaded_model.predict([[3, 9000000],[6, 2000000]])
```
